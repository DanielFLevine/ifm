```python
X_0 = torch.tensor(adata.X[:10].toarray())
X_1 = torch.tensor(adata.X[10:20].toarray())
```

```python
def generate_paths(X_0, X_1, sigma=0.1, time_points=16):
    batch_size, dim = X_0.shape
    
    # Generate time points: from 0 to 1, including both endpoints, evenly spaced
    times = torch.linspace(0, 1, steps=time_points).view(1, time_points, 1)
    
    # Expand times and inputs to broadcast across batches and dimensions
    times_expanded = times.expand(batch_size, time_points, dim)
    
    # Linear interpolation: tX_1 + (1-t)X_0 = X_0 + t(X_1 - X_0)
    path_means = X_0.unsqueeze(1) + times_expanded * (X_1 - X_0).unsqueeze(1)
    
    # Initialize paths with means (ensures exact start at X_0 and end at X_1)
    paths = path_means.clone()
    
    # Determine where X_0 or X_1 is non-zero
    non_zero_mask = ((X_0 != 0).unsqueeze(1) | (X_1 != 0).unsqueeze(1)).expand(-1, time_points, -1)

    # Gaussian noise: zero mean, sigma standard deviation, but not for the first and last time points
    if time_points > 2:
        noise = sigma * torch.randn_like(path_means[:, 1:-1])
        # Apply noise only where non_zero_mask is True, and only to intermediate points
        paths[:, 1:-1][non_zero_mask[:, 1:-1]] += noise[non_zero_mask[:, 1:-1]]

    return paths
    
```
